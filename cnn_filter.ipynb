```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def drawimg(img,title = 'Image'):
    plt.imshow(img,cmap='gray')
    plt.axis('off')
    plt.title(title + str(img.shape))
    plt.show()
        
```

```python
def convolution(img,img_filter):
    
    W = img.shape[0]
    
    H = img.shape[1]
    
    F = img_filter.shape[0]
    
    new_img = np.zeros((W-F+1,H-F+1))
    
    for row in range(W-F+1):
        
        for col in range(W-F+1):
            
            for i in range(F):
                
                for j in range(F):
                    
                    new_img[row][col] += img[row+i][col+j] * img_filter[i][j]
                
                if new_img[row][col] > 255:
                    
                    new_img[row][col] = 255
                
                if new_img[row][col] < 0:
                    
                    new_img[row][col] = 0
    return new_img
    
```

```python
img = cv2.imread("./cnn_image2.webp")

img = cv2.resize(img,(100,100))
img_ = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
drawimg(img_)
```

```python
blur_filter = np.ones((3,3))
blur_filter/=5
```

```python
blur_filter
```

```python
blurred_img = convolution(img_ , blur_filter)
```

```python
drawimg(blurred_img)
```

```python
edge_filter = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
edge_filter = edge_filter.astype('float64')
edge_filter/=0.5
img_edges = convolution(img_ , edge_filter)
```

```python
drawimg(img_edges)
```
